````python
import pandas as pd
````

````python
df = pd.read_csv('winequality-red.csv')
````

````python
df.shape
````

````python
df.head()
````

````python
# for binary classification, we can convert the quality column into a binary variable(0/1)
df['quality'] = df['quality'].apply(lambda x: 1 if x>=5.5 else 0)
df['quality'].value_counts()
# 0- bad wine, 1- good wine
````

````python
df
````

````python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
````

````python
print(X.shape, y.shape)
````

- normalising the data
- brings all features to similar scale. 
- z = x-mean/std


````python
mean = X.mean(axis = 0)
X -= mean
std = X.std(axis = 0)
X /= std
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
````

````python
model = Sequential()
model.add(Dense(8, activation='relu', input_shape=(X.shape[1],)))
# 8 neurons in this layer. This is the first hidden layer
# input_dim=len(X[0, :])
# Number of input features
# X.shape[0] = 1599 вҶ’ number of samples
# X.shape[1] = 11 вҶ’ number of features
# ReLU is used for hidden layers because:
# Faster training
# Avoids vanishing gradient
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))  # output layer
````

````python
model.summary()
````

### 1пёҸвғЈ Output Shape

* `(None, 8)` вҶ’ First layer has **8 neurons**
* `(None, 4)` вҶ’ Second layer has **4 neurons**
* `(None, 1)` вҶ’ Output layer has **1 neuron**
* `None` = batch size (number of samples, not fixed)



### 2пёҸвғЈ Param # (How Parameters Are Calculated)

**Formula for Dense layer:**

[
\text{Parameters} = (inputs \times neurons) + neurons
]

вң” First layer:
[
(11 \times 8) + 8 = 96
]

вң” Second layer:
[
(8 \times 4) + 4 = 36
]

вң” Output layer:
[
(4 \times 1) + 1 = 5
]



### 3пёҸвғЈ Total Parameters

[
96 + 36 + 5 = 137
]

All are trainable вҶ’ so:

* **Total params = 137**
* **Trainable params = 137**
* **Non-trainable params = 0**



### рҹ”‘ Key Concept

Each neuron has:

* One weight per input
* One bias

So:

```
Total params = (inputs Г— neurons) + biases
Biases = number of neurons
```






````python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
````

- Optimizer вҶ’ How the model learns
- Loss вҶ’ What the model tries to minimize
- Metrics вҶ’ How we check performance

````python
model.fit(X, y, epochs=100)
````

````python
predictions = model.predict(X)
print(predictions[:5])
````

````python
print(y[:5])
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
````

````python
acc = accuracy_score(y, predictions.round())
prec = precision_score(y, predictions.round())
rec = recall_score(y, predictions.round())
f1 = f1_score(y, predictions.round())
````

````python
print(acc, prec, rec, f1)
````

Accuracy вҶ’ Overall correct

Precision вҶ’ How correct are positive predictions?

Recall вҶ’ How many positives did we catch?

F1 вҶ’ Balance of precision & recall
